```python
import pandas as pd

# Load dataset
df = pd.read_csv('student_scores_100_with_grades.csv')

# Display the first few rows
print(df.head())
```

```python

# Highest and lowest marks per student
student_perf = df.groupby("Name")["Math"].agg(["min", "max"])
print("Student Performance Summary:\n", student_perf)
```

```python
import matplotlib.pyplot as plt

# Plot grade distribution
df["Grade"].value_counts().plot(kind="bar", color="skyblue", edgecolor="black")
plt.title("Grade Distribution")
plt.xlabel("Grades")
plt.ylabel("Number of Students")
plt.show()

grade_counts = df["Grade"].value_counts()

# Pie chart visualization
plt.figure(figsize=(6, 6))
plt.pie(grade_counts, labels=grade_counts.index, autopct="%1.1f%%", startangle=140, colors=["#ff9999", "#66b3ff", "#99ff99", "#ffcc99"])
plt.title("Grade Distribution")
plt.show()
```
